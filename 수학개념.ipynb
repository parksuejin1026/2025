```python
import matplotlib.pyplot as plt
import numpy as np

# 지수함수 그래프
x = np.linspace(-3, 3, 100) # x값의 범위를 -3부터 3까지 설정
y_exp = 2**x
plt.plot(x, y_exp, label="y = 2^x")
plt.title("Exponential Function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# 로그함수 그래프 (예시)
x_log = np.linspace(0.1, 10, 100) # x는 양수여야 함
y_log = np.log2(x_log)
plt.plot(x_log, y_log, label="y = log_2(x)")
plt.title("Logarithmic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 삼각함수 그래프
x = np.linspace(-2 * np.pi, 2 * np.pi, 200) # x값 범위를 -2파이부터 2파이까지 설정
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="y = sin(x)")
plt.plot(x, y_cos, label="y = cos(x)")
plt.axhline(0, color='black', linewidth=0.5) # x축 추가
plt.axvline(0, color='black', linewidth=0.5) # y축 추가
plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def arithmetic_sequence(a1, d, n):
    """등차수열의 n번째 항까지 계산"""
    return [a1 + (i-1)*d for i in range(1, n+1)]

def geometric_sequence(a1, r, n):
    """등비수열의 n번째 항까지 계산"""
    return [a1 * r**(i-1) for i in range(1, n+1)]

# 등차수열 예시: 첫째항 2, 공차 3
terms_arith = arithmetic_sequence(2, 3, 10)
print("등차수열 10개 항:", terms_arith)

# 등비수열 예시: 첫째항 3, 공비 2
terms_geom = geometric_sequence(3, 2, 10)
print("등비수열 10개 항:", terms_geom)

# 시각화
plt.plot(terms_arith, 'o-', label="Arithmetic Sequence")
plt.plot(terms_geom, 'o-', label="Geometric Sequence")
plt.title("Sequences")
plt.xlabel("Term Number")
plt.ylabel("Value")
plt.legend()
plt.show()
```
