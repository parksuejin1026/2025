```python
import matplotlib.pyplot as plt
import numpy as np

# 1. x 값 생성 : -10에서 10까지 0.1간격으로
x = np.arange(-10, 10, 0.1)

# 2. y 값 계산 : y = x^2
y = x**2

# 3. 그래프 그리기
plt.plot(x,y)

# 4. 그래프 제목 및 축 라벨 설정
plt.title("$y = x^2$ Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# 5. 그리드 표시
plt.grid(True)

# 6. 그래프 보여주기
plt.show()
```
